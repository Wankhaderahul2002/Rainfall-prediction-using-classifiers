```python
df[(df.Evaporation>uv3)]
```

```python
df.Evaporation[(df.Evaporation> 1.2*uv3)] = 1.2*uv3
```

```python
lv = np.percentile(df['Humidity9am'],[1])[0]
```

```python
df[(df.Humidity9am< lv)]
```

```python
df.Humidity9am[(df.Humidity9am < 0.7* lv)] = 0.7*lv
```

```python
lv2 = np.percentile(df['Pressure9am'],[1])[0]
```

```python
df[(df.Pressure9am< lv2)]
```

```python
df.Pressure9am[(df.Pressure9am < 0.8* lv)] = 0.8*lv
```

```python
uv4 = np.percentile(df['Temp3pm'],[99])[0]
```

```python
df[(df.Temp3pm>uv4)]
```

```python
df.Temp3pm[(df.Temp3pm> 1.5*uv4)] = 1.5*uv4
```

* We have many columns for data of 9 am and 3 pm. Rather than finding avg of both, we can find out the difference between the two so that we can get unit change of the parameter. This would be a more effective parameter compared to average.

```python
df['Humidity_Change'] = df.Humidity3pm - df.Humidity9am
df['Pressure_Change'] = df.Pressure3pm - df.Pressure9am
df['Cloud_Change'] = df.Cloud3pm - df.Cloud9am
df['Temp_Change'] = df.Temp3pm - df.Temp9am
```
